# Matplotlib APIs
Import NumPy, Matplotlib, and Pandas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

Generate a y=sin(x) 

```python
x = np.linspace(0,10,200)
y = np.sin(x)
```

Prepare table's labels and title before displaying.

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sin(X)')
plt.plot(x,y)
```

Read CSV file to Pandas' DataFrame. <br>
Select first column (X) and second column (Y).

```python
df_name = './data/data_1d.csv'
A = pd.read_csv(df_name,header=None).as_matrix()
x = A[:,0]
y = A[:,1]
```

### Scatter Plot
Prepare table's labels and title before displaying scatter plot.

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data File ('+df_name+')')
plt.scatter(x,y)
```

```python
x_line = np.linspace(0,100,30)
y_line = 2*x_line+1
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('A line fitted to data points in ('+df_name+')')
plt.scatter(x,y)
plt.plot(x_line,y_line,c='r',marker='*')
```

```python
x = np.random.randn(100)
y = np.random.randn(100)
```

## Histogram
Displays distribution of values.

```python
hist_data = np.random.randn(10000)
```

```python
plt.title('Histogram/Distribution of X values')
plt.xlabel('X')
plt.ylabel('Frequency(X)')
plt.hist(hist_data,bins=100)
```

Let's define two distributions (e.g., behaviour of two age groups)

```python
hist_data_l = np.random.randn(4000)-2
hist_data_r = np.random.randn(4000)+2
hist_data =  np.append(hist_data_l,hist_data_r)
```

```python
plt.title('Histogram/Distribution of X values')
plt.xlabel('X')
plt.ylabel('Frequency(X)')
plt.hist(hist_data,bins=100)
```
